```python
# Import necessary libraries for numerical operations, plotting and statistics

import numpy as np
import matplotlib.pyplot as plt
import statistics as st
```

```python
# Load v_rest_sst.csv and v_rest_pvalb.csv as NumPy arrays

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest1 = np.loadtxt(url, delimiter=',')

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest2 = np.loadtxt(url, delimiter=',')
```

```python
# Combine the two arrays into a single 2x13 matrix called v_rest

v_rest = np.vstack((v_rest1, v_rest2))
```

```python
# Print the size of v_rest

print(v_rest.size)
```

```python
# Return the first row of v_rest

v_rest[0]
```

```python
# Calculate and return the sum of the third column of v_rest

v_rest[:, 2].sum()
```

```python
# Return the second-to-last value from the second row using negative indexing

v_rest[1, -2]
```

```python
# Create a histogram of the values in the first row of v_rest and label the axes

plt.hist(v_rest[0])
plt.xlabel('Membrane Potential (mV)')
plt.ylabel('Frequency')
plt.title('Histogram of First Row of v_rest')
plt.show()
```

```python
# Calculate the mean of each row in v_rest using the st.mean() function


v_rest_mean_sst = st.mean(v_rest[0])
print(v_rest_mean_sst)

v_rest_mean_pvalb = st.mean(v_rest[1])
print(v_rest_mean_pvalb)
```

# Describe the difference in means between the two rows. Which is larger?

Since a more positive (less negative value) is larger, -67.73 is greater than -69.97, so the SST neurons have the larger (less negative) mean resting membrane potential.

```python
# Import the scipy.stats package to perform a t-test

import scipy.stats as stats
```

```python
# Perform an independent samples t-test comparing the first and second rows of v_rest

stats.ttest_ind(v_rest[0], v_rest[1])
```

# Describe The Results of the t-Test

An ind. samples t-test was conducted to compare the mean resting membrane potentials of SST and PVALB neurons.

The t-statistic was 5.84, and the p-value was 7.16 × 10⁻⁹.

Because the p-value (7.16 × 10⁻⁹) is less than 0.05 (5%), the difference in mean resting membrane potentials between SST and PVALB neurons is statistically significant.

Therefore, it is highly unlikely that the observed difference in means occurred by chance.
