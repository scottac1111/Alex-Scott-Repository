```python
#make a code cell that imports numpy as np, matplotlib.pyplot as plt, and scipy.stats as stats

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
#make a code cell that imports your v_rest_sst.csv and v_rest_pvalb.csv variables as numpy arrays called v_rest_sst and v_rest_pvalb respectively


url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_sst.csv'
v_rest_sst = np.loadtxt (url, delimiter = ',')

url = 'https://raw.githubusercontent.com/hallockh/neur_265_spring2026/refs/heads/main/data/v_rest_pvalb.csv'
v_rest_pvalb = np.loadtxt (url, delimiter = ',')
```

```python
#Find the means

import numpy as np

mean_sst = np.mean(v_rest_sst)
mean_pvalb = np.mean(v_rest_pvalb)

print("Mean v_rest_sst:", mean_sst)
print("Mean v_rest_pvalb:", mean_pvalb)
```

```python
#perform t-test

from scipy import stats

t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)

print("t-statistic:", t_stat)
print("p-value:", p_val)
```

```python
#plot a histogram of v_rest_sst in one subplot and a histogram of v_rest_pvalb in another subplot if the p-value of the test is <.05.

from scipy import stats
import matplotlib.pyplot as plt

# Standard independent t-test
t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)

print("p-value:", p_val)

if p_val < 0.05:
    fig, ax = plt.subplots(1, 2)

    ax[0].hist(v_rest_sst, bins=15)



    ax[1].hist(v_rest_pvalb, bins=15)



    plt.tight_layout()
    plt.show()
else:
    print("p >= 0.05 — no histograms plotted.")
```

```python
#add x and y-axes

from scipy import stats
import matplotlib.pyplot as plt

# Standard independent t-test
t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)

print("p-value:", p_val)

if p_val < 0.05:
    fig, ax = plt.subplots(1, 2)

    ax[0].hist(v_rest_sst, bins=15)
    ax[0].set_title("v_rest_sst")
    ax[0].set_xlabel("Resting Potential")
    ax[0].set_ylabel("Frequency")

    ax[1].hist(v_rest_pvalb, bins=15)
    ax[1].set_title("v_rest_pvalb")
    ax[1].set_xlabel("Resting Potential")
    ax[1].set_ylabel("Frequency")

    plt.tight_layout()
    plt.show()
else:
    print("p >= 0.05 — no histograms plotted.")
```

```python
#make a code cell that defines a function called vrestHist()

def vrestHist():
    from scipy import stats
    import matplotlib.pyplot as plt

    t_stat, p_val = stats.ttest_ind(v_rest_sst, v_rest_pvalb)
    print("p-value:", p_val)

    if p_val < 0.05:
        fig, ax = plt.subplots(1, 2)

        ax[0].hist(v_rest_sst)
        ax[0].set_title("v_rest_sst")

        ax[1].hist(v_rest_pvalb)
        ax[1].set_title("v_rest_pvalb")

        plt.show()
    else:
        print("p >= 0.05, no histograms plotted.")
```

```python
#make a code cell that calls vrestHist()

vrestHist()
```

```python
#make a scatterplot with v_rest_sst on the x-axis and v_rest_pvalb on the y-axis

import matplotlib.pyplot as plt

plt.figure()

for sst, pvalb in zip(v_rest_sst, v_rest_pvalb):

    if sst > -60 and pvalb > -60:
        color = "red"
    elif sst > -60 and pvalb < -60:
        color = "green"
    elif sst < -60 and pvalb < -60:
        color = "blue"
    else:  # sst < -60 and pvalb > -60
        color = "purple"

    plt.scatter(sst, pvalb, color=color)

plt.xlabel("v_rest_sst")
plt.ylabel("v_rest_pvalb")
plt.title("SST vs PVALB Resting Potential")
plt.show()
```
