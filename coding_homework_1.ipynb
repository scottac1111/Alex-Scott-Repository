```python
pandas_data1 = pd.read_csv("v_rest_sst.csv")
pandas_data2 = pd.read_csv("v_rest_pvalb.csv")
```

```python
# Convert Pandas DataFrames into Python lists

data1 = pandas_data1.values.tolist()
data2 = pandas_data2.values.tolist()
```

```python
# Flatten the lists to remove unnecessary brackets

data1 = np.squeeze(data1)
data2 = np.squeeze(data2)
```

```python
# Create a histogram of data1
plt.hist(data1)

# Display the plot
plt.show()
```

```python
plt.hist(data1)

# Label the x and y axes
plt.xlabel("Values")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

```python
#Create a histogram of data2
plt.hist(data2)

# Label the x and y axes
plt.xlabel("Values")
plt.ylabel("Frequency")

# Display the plot
plt.show()
```

One noticeable difference between the two histograms is that the x-axes and y-axes are scaled differently. E.g. the x-axis of data 1 starts from -90 to -50 while data 2 starts from -85 to -45. The peak of data 1 is centered around -70 and, similarly, the peak of dat 2 is centered around a value of -70. However, the histogram of data 1 looks to be more skewed left, while data 2's histogram appears skewed right. The cells in data 1 appear to have slightly more negative resting membrane potentials on average compared to data 2, as their distribution is centered a bit further to the left. In contrast, data 2 appears slightly shifted toward less negative (higher) membrane potential values. This suggests that the cells in data 2 may have a somewhat higher resting membrane potential than the cells in data 1.
